```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from shennong.utils import symbol, trading_days
from shennong.stk import bar
import statsmodels.api as sm
```

```python
sym_list = symbol.load(region='cn',product='ashare')
date_line = trading_days.load('2014-01-02', '2020-08-27', region='cn', product='ashare')
```

```python
daily_stocks = bar.load(start_datetime = date_line[0], end_datetime = date_line[-1],
                     region='cn', product='ashare', freq='1day',symbol_list = sym_list,
                     key_list='raw_wangbo_basic_cn_ashare_1day',
                     load_root='default', verbose=True, restoration='backward_restoration')
```

```python
daily_stocks
```

```python
#volume = daily_stocks.loc[:,:,'raw_wangbo_turnover'].to_pandas().cumsum(axis=1)
volume = daily_stocks.loc[:,:,'raw_wangbo_turnover'].to_pandas()
volume = volume.T
volume.index = volume.index.strftime("%Y-%m-%d")
```

```python
close = daily_stocks.loc[:,:,'raw_wangbo_close'].to_pandas()
```

```python
preclose = daily_stocks.loc[:,:,'raw_wangbo_preclose'].to_pandas()
```

```python
daily_ret = close/preclose - 1
```

```python
daily_ret_df = daily_ret.T
```

```python
daily_ret_df.index = daily_ret_df.index.strftime("%Y-%m-%d")
```

```python
#
daily_ret_df = daily_ret_df * volume
```

```python
def get_sum_ret_list(daily_ret_df):
    sum_ret_series = pd.Series()
    for i in range(daily_ret_df.shape[1]):
        sum_ret_series = pd.concat([sum_ret_series,pd.Series(daily_ret_df.iloc[:,i].sum())])
    return sum_ret_series
```

```python
sum_ret = get_sum_ret_list(volume)
```

```python
sum_ret.index = daily_ret_df.columns
```

```python
sorted_sum_ret= sum_ret.sort_values(ascending=False)
```

```python
sorted_sum_ret
```

```python
array1 = np.array_split(sorted_sum_ret,10)
```

```python
array1
```

```python
ind_array = [i.index for i in array1]
```

```python
first_line = daily_ret_df.loc[:,ind_array[0]].apply(np.nanmean,axis=1).cumsum()
second_line = daily_ret_df.loc[:,ind_array[1]].apply(np.nanmean,axis=1).cumsum()
third_line = daily_ret_df.loc[:,ind_array[2]].apply(np.nanmean,axis=1).cumsum()
fourth_line = daily_ret_df.loc[:,ind_array[3]].apply(np.nanmean,axis=1).cumsum()
fifth_line = daily_ret_df.loc[:,ind_array[4]].apply(np.nanmean,axis=1).cumsum()
sixth_line = daily_ret_df.loc[:,ind_array[5]].apply(np.nanmean,axis=1).cumsum()
seventh_line = daily_ret_df.loc[:,ind_array[6]].apply(np.nanmean,axis=1).cumsum()
eighth_line = daily_ret_df.loc[:,ind_array[7]].apply(np.nanmean,axis=1).cumsum()
ninth_line = daily_ret_df.loc[:,ind_array[8]].apply(np.nanmean,axis=1).cumsum()
tenth_line = daily_ret_df.loc[:,ind_array[9]].apply(np.nanmean,axis=1).cumsum()
```

```python
plt.figure(figsize=(18,8))
plt.plot(volume.index,first_line,label = "1")
plt.plot(volume.index,second_line,label = "2")
plt.plot(volume.index,third_line,label = "3")
plt.plot(volume.index,fourth_line,label = "4")
plt.plot(volume.index,fifth_line,label = "5")
plt.plot(volume.index,sixth_line,label = "6")
plt.plot(volume.index,seventh_line,label = "7")
plt.plot(volume.index,eighth_line,label = "8")
plt.plot(volume.index,ninth_line,label = "9")
plt.plot(volume.index,tenth_line,label = "10")
plt.xlabel("Date")
plt.ylabel("Return")
plt.xticks([volume.index[i] for i in range(len(volume.index)) if i%50 == 0],rotation = 45)
plt.title("Ts_return")
plt.legend()
```

```python
"""

"""
def compute_corr(df1,df2):
    corr_list = []
    for i in range(df1.shape[0]):
        corr_list.append(df1.iloc[i,:].corr(df2.iloc[i,:])) 
    return corr_list
```
